```python
from pydantic import BaseModel

class User(BaseModel):
    name: str
    age: int
    email: str

user_data = {"name": "Charlie", "age": 40, "email": "charlie@example.com"}
user_instance = User(**user_data)

print(user_instance.name)
print(user_instance.email)
```
