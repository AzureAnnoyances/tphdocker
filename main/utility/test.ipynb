```python
import numpy as np
class Oitems:
    def __init__(self, i, xy, z_ffb, z_grd):
        self.i:int=i
        self.xy:np.ndarray=xy
        self.x = xy[0]
        self.y = xy[1]
        self.z_ffb:float=z_ffb
        self.z_grd:float=z_grd
        self.detected_trunk:bool=False
        self.detected_crown:bool=False
        pass
    
    def asdict(self):
        return self.__dict__
    
```

```python
list_dict = []
for i in range(3):
    arr = np.array((1*2,5*i))
    list_dict.append(Oitems(i, arr, 0.2*i, 0.5*i))

import pandas as pd
dfs=[]
for item in list_dict:
    dfs.append(item.asdict())
combined_df = pd.DataFrame(dfs, columns=["i","x","y","detected_trunk","detected_crown"])
print(combined_df)
```
